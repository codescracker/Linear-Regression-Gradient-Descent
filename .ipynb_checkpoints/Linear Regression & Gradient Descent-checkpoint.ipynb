# Generate artificial data

```python
X = 2* np.random.rand(100, 1)
```

```python
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
X_b = np.c_[np.ones((100, 1)), X]
```

# Math solution for theta_best

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

```python
X_new = np.array([[0],
                  [2]])
```

```python
X_new_b = np.c_[np.ones((2, 1)), X_new]
```

```python
y_pred =  X_new_b.dot(theta_best)
```

```python
y_pred
```

```python
plt.plot(X, y, 'b.')
plt.plot(X_new, y_pred, 'r-')
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

# Batch Gradient Descent Solution for fit theta_best

```python
eta = 0.1 ## this is the learning rate(step length)
n_iteration = 1000
m = 100 ## the number of instance in training set

theta = np.random.randn(2,1)
BGD_thetas = [theta]

for i in range(0,n_iteration):
    gradients = 2*1.0/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta*gradients
    BGD_thetas.append(theta)
```

```python
theta
```

# Visualize the fitting process of BGD

```python
def plotBGD(X,y,eta, n_iteration ,ax):
    m = len(X)
    X_b = np.c_[np.ones((m,1)), X]
    
    theta = np.random.randn(2,1)
    
    ax.plot(X,y,'b.')
    X_new = np.array([[np.min(X)],
                      [np.max(X)]])
    X_new_b = np.c_[np.ones((2,1)),X_new]
    
    for i in range(0,n_iteration):
        if i<10:
            y_pred = X_new_b.dot(theta)
            if i==0:
                ax.plot(X_new,y_pred, 'r--')
            else:
                ax.plot(X_new,y_pred, 'b-')
        
        gradients = 2*1.0/m * X_b.T.dot(X_b.dot(theta) - y)
        theta = theta - eta*gradients
    
    ax.set_xlabel("$x_1$", fontsize=18)
    ax.set_ylabel("$y$", rotation=0, fontsize=18)
    ax.set_xlim(0,2)
    ax.set_title(r"$\eta = {}$".format(eta), fontsize=16)
```

```python
fig,(ax1, ax2, ax3)  = plt.subplots(1,3,figsize = (16,8))
plotBGD(X,y,0.02,1000, ax1)
plotBGD(X,y,0.1,1000, ax2)
plotBGD(X,y,0.5,1000, ax3)
```

# Stochastic Gradient Descent

```python
def learning_rate(t0,t1,t):
    return t0*1.0/(t+t1)
```

```python
theta = np.random.randn(2,1)
SGD_thetas = [theta]
n_epochs = 50
m = len(X)
t0,t1 = 5,50

for epoch in range(0,n_epochs):
    for i in range(0,m):
        random_idx = np.random.randint(m)
        xi = X_b[random_idx:random_idx+1]
        yi = y[random_idx:random_idx+1]
        
        gradients = (2*1.0/1) * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_rate(t0, t1, epoch*m+i)
        theta = theta - eta*gradients
        SGD_thetas.append(theta)
```

```python
theta
```

# Visualize the learning process of SGD

```python
def plotSGD(X,y,n_epoch,t0,t1,ax):
    m = len(X)
    X_b = np.c_[np.ones((m,1)), X]
    
    theta = np.random.randn(2,1)
    
    ax.plot(X,y,'b.')
    X_new = np.array([[np.min(X)],
                      [np.max(X)]])
    X_new_b = np.c_[np.ones((2,1)),X_new]
    
    for epoch in range(0,n_epochs):
        for i in range(0,m):
            if epoch ==0 and i<10:
                y_pred = X_new_b.dot(theta)
                if i==0:
                    ax.plot(X_new,y_pred, 'r--')
                else:
                    ax.plot(X_new,y_pred, 'b-')
            
            random_idx = np.random.randint(m)
            xi = X_b[random_idx:random_idx+1]
            yi = y[random_idx:random_idx+1]

            gradients = (2*1.0/1) * xi.T.dot(xi.dot(theta) - yi)
            eta = learning_rate(t0, t1, epoch*m+i)
            theta = theta - eta*gradients

    ax.set_xlabel("$x_1$", fontsize=18)
    ax.set_ylabel("$y$", rotation=0, fontsize=18)
    ax.set_xlim(0,2)
    ax.set_title("Stochastic Gradient Descent", fontsize=16)
```

```python
fig,ax  = plt.subplots(figsize = (8,8))

n_epochs = 50
m = len(X)
t0,t1 = 5,50

plotSGD(X,y,n_epochs,t0,t1,ax)
```
